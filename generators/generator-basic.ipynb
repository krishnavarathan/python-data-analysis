```python
import sys
```

```python
def num_sqrt(x):
    for i in x:
        yield i*i
```

```python
def num_sqrt2(x):
    for i in x:
        return i*i
```

```python
values = num_sqrt([1,2,3,4])
print(next(values))
print(next(values))
print(next(values))
print(next(values))
```

```python
values_2 = num_sqrt2([1,2,3,4])
print(sys.getsizeof(values_2))
```
