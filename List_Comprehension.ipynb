```python
l=[2,4,5,6,89,186]
```

```python
sq=[num*2 for num in l]
print(sq)
```

```python
sq=[num*2 for num in l if num%2==0]
print(sq)
```

```python
# sq=[num*2 if num%2==0 for num in l]
# print(sq)
```

```python
sq=[num*2 if num%2==0 else print("Even") for num in l]
print(sq)
print(type(sq))
```

```python
print([i for i in range(10)])
```

```python
# Nested for loop inside list
a=[i*j for i in range(5) for j in range(5)]
print(len(a))
print(type(a))
print(a)
```
