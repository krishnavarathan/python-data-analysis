```python
#⭐ Problem 2 – Even or Odd
print((lambda x:True if x%2==0 else False)(2))
print((lambda x:True if x%2==0 else False)(3))
```

```python
# ⭐ Problem 3 – Add Two Numbers
print((lambda x,y:x+y) (20,45))
```

```python
# ⭐ Problem 4 – Find Maximum using Max()
a = [1,2,3,4,5]
res=max(list(map(lambda x:x*3, a)))
print(res)
```

```python
#⭐ Problem 5 – Multiply by 3 List Values Using map()
print(list(map(lambda x:x*3, a)))
```

```python
# Problem 6 – Filter Even Numbers
nums = [10,15,20,25,30]
result = list(filter(lambda x: x%2==0, nums))
print(result)
```

```python
# ⭐ Problem 7 – Convert Names to Uppercase
names = ["krishna","python","data"]
up=list(map(lambda x:x.upper(), names))
print(up)
```

```python
#⭐ Problem 8 – Sort List of Tuples
data = [("A", 25), ("B", 15), ("C", 35)]
print(sorted(data))
print(sorted(data, key=lambda x:x[1]))
```

```python
# ⭐ Problem 9 – Add 10 to Each Number Using List Comprehension + Lambda
nums = [1,2,3,4]
nums=[n+10 for n in nums]
print(nums)
numx=list(map(lambda x:x+10, nums))
print(numx)
```
