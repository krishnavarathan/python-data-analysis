## pandas Series: .sort_index()

```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Pandas_Exercises/Location.csv')
location_data = data.copy()
location_data.head()
```

```python
location_data.sort_values(by='Location', ascending=False)
```

```python
location_data_sv = location_data.sort_values(by='Location', ascending=False)
location_data_sv.head()
```

```python
location_data_sv.index
```

```python
location_data_sv.index.sort_values()
```

```python
location_data_sv.head()
```

```python
location_data_sv.index = location_data_sv.index.sort_values()
```

```python
location_data_sv
```

```python
location_data_sv = location_data.sort_values(by='Location', ascending=False)
location_data_sv.head()
```

```python
location_data_sv.sort_index()
```

```python
location_data_sv
```

```python
location_data_sv = location_data_sv.sort_index(ascending=True)
location_data_sv
```
