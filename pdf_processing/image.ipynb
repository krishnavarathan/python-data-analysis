```python
import fitz
import PIL.Image
import io
import os
import re
from pdfminer.high_level import extract_pages, extract_text

provisinal_pdf = r"Data/DT20234570014_Provisinal Offer Letter.pdf"
hcl_pdf= r"C:\Users\krishna varathan\Documents\PDF_Spitting\Data\Application_form_HCLTFP1880079.pdf"
application_pdf=r"Data/DT20234570014_Application_Form.pdf"
payslip_pdf=r"Data/Jan.pdf"
llr_pdf=r"Data/KARUPPU_LLR.pdf"
```

```python
files = [provisinal_pdf,hcl_pdf,application_pdf,payslip_pdf,llr_pdf]  
files
```

```python
count=0
for file in files:
    print(file)
    text=extract_text(file)
    pattern = re.compile(r"[a-zA-Z]+\s{1}+.{1,100}")
    matches = pattern.findall(text)
    for i in matches:
        # print(i)
        count+=1
print(count)
        
```

```python
pdf = fitz.open(llr_pdf)
print(type(pdf))
print(len(pdf))
print(pdf.__sizeof__())
print(pdf[0].get_images())
print(len(pdf[0].get_images()))

for i in range(len(pdf)):
    page=pdf[i]
    images=page.get_images()
    for image in images:
        base_img=pdf.extract_image(image[0])
        image_data=base_img['image']
        img=PIL.Image.open(io.BytesIO(image_data))
        extension=base_img['ext']
        img.save(open(f"image{count}.{extension}", 'wb'))
        count+=1
```

```python
total_images=1
for file in files:
    count = 1
    file_name=os.path.basename(file)
    file_name=os.path.splitext(file_name)
    # file_name[0] # actually this is the file name without extension
    
    pdf=fitz.open(file)
    for i in range(len(pdf)):
        page=pdf[i] # Get the page from individual pdf
        images=page.get_images() # Get the images from the page, its just for counting the number of images in the page
        for image in images:
            base_img=pdf.extract_image(image[0])
            extension=base_img['ext']
            img_data=base_img['image']
            img=PIL.Image.open(io.BytesIO(img_data))
            try:
                os.mkdir(file_name[0]) # create a folder with the file name
                file_path=os.path.join(file_name[0], f"{file_name[0]}_image{count}.{extension}")
                img.save(open(file_path, 'wb'))
            except Exception as e:
                print(f"Error saving image{e}")

            total_images+=1
            count+=1
print(f"{total_images} images extracted from the pdf files and saved in the respective folders")
```
