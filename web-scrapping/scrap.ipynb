```python
from bs4 import BeautifulSoup
import cloudscraper
!pip install lxml
```

```python
# Create a cloudscraper session to bypass Cloudflare protection
scraper = cloudscraper.create_scraper(
    browser={
        'browser': 'chrome',
        'platform': 'windows',
        'desktop': True
    }
)
```

```python
# Target URL
url = 'https://www.indiafilings.com/partnership'
url
```

```python
print('Attempting to fetch the page...')
response = scraper.get(url, timeout=30)
print('Status Code:', response.status_code)
```

```python
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')
    if soup.title:
        print('Page Title:', soup.title.text)
    text = soup.get_text(strip=True)
    print('Content length:', len(text), 'characters')
```

```python
# Extract headings
print('--- Headings found ---')
for tag in soup.find_all(['h1', 'h2', 'h3']):
    text = tag.get_text(strip=True)
    if text:
        print(tag.name + ':', text)
```

```python
# Extract headings
print('--- Headings found ---')
for tag in soup.find_all(('li')):
    text = tag.get_text(strip=True)
    if text:
        print(tag.name + ':', text)
```

```python
# Extract first few paragraphs
print('--- First 5 paragraphs ---')
for p in soup.find_all('p')[:5]:
    text = p.get_text(strip=True)
    if text:
        print('p:', text[:200])
```

```python
# Extract first few links
print('--- First 10 links ---')
for link in soup.find_all('a')[:10]:
    href = link.get('href')
    text = link.get_text(strip=True)
    if href and text:
        print('Link:', text, '->', href)
```
