```python
filename = '/content/This is python text book.txt'
print(type(filename))
```

```python
file = open(filename, mode = 'r')
print(type(file))
```

```python
text = file.read()
print(type(text))
```

```python
print(file.closed)
file.close()
print(file.closed)
print(text)
```

```python
print(len(text))
print(list(text))

# for i in text:
#   print(i)
```

```python
digits = ""
final =[]
for i in text:
  if i != " ":
    digits+=i
  else:
    final.append(int(digits))
    digits =""
if digits:
  final.append(int(digits))
print(final)
```

```python
import pandas as pd
final_array = pd.Series(final, name="numbers", index=[2,1,3,4])
print(final_array.to_frame())
print(final_array[2])
print(final_array.index[2])
```
