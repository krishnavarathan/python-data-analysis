**Minimal and Maximal Values in NumPy - Exercise #1**
Considering the dimensionality and content of the provided NumPy arrays, display the following in the suggested order:

Call the np.amax() function and pass array_2D as input.

Call the np.amax() function along the second axis and pass array_3D as input.

```python
import numpy as np

np.set_printoptions(suppress = True, linewidth = 150, precision = 2)

array_1D = np.array([10,11,12,13,14])
array_2D = np.array([[20,30,40,50,60], [43,54,65,76,87], [11,22,33,44,55]])
array_3D = np.array([[[1,2,3,4,5], [11,21,31,41,51]], [[11,12,13,14,15], [51,52,53,54,5]]])

#1.
print(np.max(array_1D))
#2.
print(np.amax(array_2D))
#3.
print(np.amax(array_1D, axis = 0))
#4.
print(np.amax(array_2D, axis = 1))
#5.
print(np.amax(array_3D, axis = 2))
#6.
print(np.maximum.reduce(array_1D))
#7.
print(np.maximum.reduce(array_3D))
```

**Statistical Order Functions in NumPy - Exercise #2**
Consider the dimensionality and content of the provided NumPy arrays.

1. Sort the values in array_1D in ascending order, then:

- Use np.percentile() to compute and display the 30th percentile of array_1D.

- Use np.quantile() to compute and display the 66th quantile of array_1D.

- Use np.quantile() to compute and display the median of array_1D.

2. Repeat the same 4 steps for array_2D.

**Statistical Functions in NumPy - Exercise #1**
Based on the dimensionality and content of the provided two-dimensional array array_2D, compute and display the following statistical values:

1. Mean.

2. Median.

3. Column-wise mean.

4. Variance.

5. Standard deviation.

6. Covariance (treating each row as a variable and each column as an observation).

7. Correlation (treating each row as a variable and each column as an observation).

**Histograms in NumPy - Exercise #1**
Based on the dimensionality and content of the provided two-dimensional array array_2D, compute and store the following histogram data:

1. A two-dimensional histogram using:

the first row of array_2D as the x values,

the second row of array_2D as the y values,

and 4 bins for each dimension.
Store the result in a variable named result_2d.

2. A multi-dimensional histogram with 4 bins per dimension, treating the columns of array_2D as independent observations in 3-dimensional space.
Store the result in a variable named result_dd.

```python
array_2D = np.array([[20,30,40,50,60], [43,54,65,76,87], [11,22,33,44,55]])

#1.
result_2d = np.histogram2d(array_2D[0], array_2D[1], bins=4)
#2.
result_dd = np.histogramdd(array_2D.transpose(), bins=4)

result_2d
result_dd
```

```python
import matplotlib.pyplot as plt
plt.hist(array_2D, bins=4)
plt.show()
```
